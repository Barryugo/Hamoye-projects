```python
# importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

This is a quiz on linear regression which requires numerical data for the model and every data in here are numerical except the 'date' columns. The column will be dropped soon.

```python
energy1 = energy.drop(columns=['date', 'lights'])
```

<h4> Normalizing the dataset using MinMax scaler

```python
# Normalizing
scalar = MinMaxScaler()
normalized_energy = pd.DataFrame(
    scalar.fit_transform(energy1), columns=energy1.columns)

# Splitting to feature and target variables
energy_x = normalized_energy.drop('Appliances', axis=1)
target_y = normalized_energy['Appliances']


```

<h4>Fitting a linear model on the relationship between T2 and T6

```python
T2_T6 = normalized_energy[['T2', 'T6']]

# Why is this reshapep... maybe because it doesnot accept one dimensional
T2_T6x = T2_T6['T2'].values.reshape(-1, 1)
T2_T6y = T2_T6['T6'].values.reshape(-1, 1)

# splitting to train and test set
x_train, x_test, y_train, y_test = train_test_split(T2_T6x, T2_T6y,
                                                    test_size=0.30, random_state=42)
```

```python
# The linear model
T2_T6_linear_model = linear_model.LinearRegression()
T2_T6_linear_model.fit(x_train, y_train)
T2_T6_linear_model_pred = T2_T6_linear_model.predict(x_test)
T2_T6_r2_score = r2_score(y_test, T2_T6_linear_model_pred)
r2 = round(T2_T6_r2_score, 2)
display('r2_score of...', r2)
```

<h3>Training this model on the main dataset

```python
# splitting to train and test sets
train_x, test_x, train_y, test_y = train_test_split(energy_x, target_y,
                                                    test_size=0.30, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(train_x, train_y)
pred_values = linear_model.predict(test_x)
```

```python

mae = mean_absolute_error(test_y, pred_values)
print('The MAE is', (round(mae, 2)))
```

```python
rss = np.sum(np.square(test_y - pred_values))
print('The RSS is', (round(rss, 2)))
```

```python

rmse = np.sqrt(mean_squared_error(test_y, pred_values))
print('The RMSE is', (round(rmse, 3)))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(test_y, pred_values)
print('R square, also known as coefficient of determination is', (round(r2, 2)))
```

```python
# comparing the effect of the regularisation

def get_weight_energy(model, feat, col_name):
    '''This function returns the weight of every feature'''
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_energy = pd.DataFrame(weights).reset_index()
    weights_energy.columns = ['Features', col_name]
    weights_energy[col_name].round(3)
    return weights_energy


linear_model_weight = get_weight_energy(
    linear_model, train_x, 'Linear_Model_Weight')
linear_model_weight
```

rv1 has the highest weight and rv2 has the lowest

<h4>Training a Ridge Regression model

```python
ridge = Ridge(alpha=0.4)
ridge.fit(train_x, train_y)
ridge_prediction = ridge.predict(test_x)
rmse = np.sqrt(mean_squared_error(test_y, ridge_prediction))
print('The RMSE is', (round(rmse, 3)))
```

The RMSE remains the same after fitting with the ridge regresion.

<h4>Training a Lasso Regression model

```python
lasso = Lasso(alpha=0.001)
lasso.fit(train_x, train_y)
lasso_prediction = lasso.predict(test_x)
linear_model_weight = get_weight_energy(lasso, train_x, 'Lasso_Weight')
linear_model_weight.sort_values(by='Lasso_Weight')
```

RH_out, RH_8, Windspeed and RH_1 has a non-zero feature, so there are just 4 of them.

```python

rmse = np.sqrt(mean_squared_error(test_y, lasso_prediction))
print('The RMSE is', (round(rmse, 3)))
```
